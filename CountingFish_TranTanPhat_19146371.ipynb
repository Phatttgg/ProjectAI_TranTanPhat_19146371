## **Khai báo thư viện và Training**

```python
from detecto import core, utils, visualize
import matplotlib.pyplot as plt
import matplotlib.image as img

# mention you dataset path
dataset = core.Dataset('/content/drive/MyDrive/dataset')
# mention you object label here
model = core.Model(['fish'])


#train data
model.fit(dataset, epochs=15, verbose=1)
```

## **Lưu model**

```python
model.save('/content/drive/MyDrive/dataset/fish_model.pth')
```

## **Test thử ảnh**

```python
from detecto import core, utils, visualize
import numpy as np
import seaborn as sns
from torchvision import transforms as tf
from detecto.visualize import show_labeled_image, plot_prediction_grid
import collections

image = utils.read_image('/content/drive/MyDrive/dataset/fish22.jpg') 
labels, boxes, scores = model.predict(image)
filtr_ind=np.where(scores>0.798)
filtr_scr=scores[filtr_ind]
filtr_boxes=boxes[filtr_ind]
num_list = list(filtr_ind[0])
filtr_labels = []
for i in num_list:
  filtr_labels.append(labels[i] + "#" + str(i+1))
show_labeled_image(image, filtr_boxes, filtr_labels)
# print(labels) 
# print(boxes)
print(scores)

c = collections.Counter(filtr_labels)
print(c)
print('So luong ca:',len(c))
```
